```python
import re
```

```python
def preprocessing(text):
    text = re.sub(r'\n+', ' ', text)                 # Remove newlines
    text = re.sub(r'\"?\\?&?gt;?', '', text)         # Remove quotes
    text = re.sub(r'\*', '', text)                   # Remove asterisks
    text = re.sub('&amp;#x200B;', '', text)          # Remove bullet points
    text = re.sub(r'\[.*?\]\(.*?\)', '', text)       # Remove links
    text = re.sub('~', '', text)                     # Remove strikethroughs
    text = re.sub('&lt;', '', text)                  # Remove spoilers
    text = re.sub(r'!(.*?)!', r'\1', text)
    text = re.sub('`', '', text)                     # Remove code
    text = re.sub(r'\^\((.*?)\)', r'\1', text)       # Remove superscript
    text = re.sub(r'\|', ' ', text)                  # Remove table pipes
    text = re.sub(':-', '', text)                    # Remove table formatting
    text = re.sub('#', '', text)                     # Remove headings
    text = re.sub(r'<.*?>', '', text)                # Remove HTML tags
    text = re.sub(r'\b\d+\b', '', text)  # Remove numbers
    text = text.strip()
    text = re.sub(r'\s\s+', ' ', text)
    text = text.lower()                              # Convert text to lowercase
    return text
```

```python
df['post_1'] = df['post'].apply(lambda text:preprocessing(text))
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')

```

```python
from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words('english'))

def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df['post_remove_stopwords'] = df['post_1'].apply(lambda text: remove_stopwords(text))
```

```python
import string

PUNCTUATION = string.punctuation

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', PUNCTUATION))

df['post_remove_punctuation'] = df['post_remove_stopwords'].apply(lambda text: remove_punctuation(text))
```
