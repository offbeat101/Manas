```python

import pandas as pd

```

```python
folder= '/content/drive/MyDrive/nlpdataset'
```

```python
df = pd.DataFrame()
```

```python

import os

files = os.listdir(folder)

for file in files:
  df_temp = pd.read_csv(os.path.join(folder, file))
  df = pd.concat([df, df_temp], ignore_index=True)
```

```python
df
```

```python

df = df[['subreddit', 'post']]
```

```python
df
```

```python
df['subreddit'].unique()
```

```python
import seaborn as sns


sns.countplot(y="subreddit", data=df)
```

```python

df = df.drop(df[df['subreddit'] == 'depression'].sample(33000).index)
df = df.drop(df[df['subreddit'] == 'suicidewatch'].sample(30000).index)
df = df.drop(df[df['subreddit'] == 'mentalhealth'].sample(20000).index)
df = df.drop(df[df['subreddit'] == 'anxiety'].sample(10000).index)
```

```python
import seaborn as sns


sns.countplot(y="subreddit", data=df)
```
